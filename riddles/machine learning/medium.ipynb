```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('MlMediumTrainingData.csv')

# Extract features (x, y coordinates) and labels
X = data[['x_', 'y_']].values
y = data['class'].values
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(data['x_'], data['y_'], c=data['class'], cmap='viridis')
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.title('Plot of CSV Data')
plt.colorbar(label='Label')
plt.show()
```

```python
from sklearn.cluster import DBSCAN
eps = 0.1  # DBSCAN epsilon parameter
min_samples = 5  # Minimum number of samples required to form a cluster
algorithm = DBSCAN(eps=eps, min_samples=min_samples)
algorithm.fit(X)
predictions = algorithm.labels_
```

```python
# Plotting
plt.figure(figsize=(8, 6))
plt.scatter(X[predictions == 0, 0], X[predictions == 0, 1], c='blue', label='Connected Shape')
plt.scatter(X[predictions == -1, 0], X[predictions == -1, 1], c='red', label='Not Connected')
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.title('Connected Components Detection')
plt.legend()
plt.show()
```
